```python
#Pandas is the library and its contains functions related to database
import pandas as pd
```

```python
#Load Original Dataset for ML
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
#Categorical data have to be preprocess to change the numerical data.
#if categorical data is nominal data type we should use 'One Hot Encoding' concept to preprocess the data.
#To reduce the duplication, its required to drop the first column to enhance the process in 'One Hot Encoding'concept
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
#Split Input Data from Original Data
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
#Split Output Data from Original Data
dependent=dataset[['charges']]
```

```python
#Split the Train Data and Test Data and load in different variables
from sklearn.model_selection import train_test_split as TTS
x_train,x_test,y_train,y_test=TTS (independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
Grid_Dictionary ={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'C':[10,100,1000,2000,3000]}
Grid=GridSearchCV(SVR(),Grid_Dictionary,refit=True,verbose=3,n_jobs=-1)
Grid.fit(x_train,y_train)
```

```python
#Print best parameter after tuning
print(Grid.best_params_)
```

```python
re=Grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
#Evaluvating the created model by using Test value as given input for the created model
y_predict=Grid.predict(x_test)
y_predict
```

```python
#compare the y_predit value with original y_test value in R2 evalution metrics to evaluate the model
from sklearn.metrics import r2_score
Evaluvation_Score=r2_score(y_test,y_predict)
Evaluvation_Score
```
